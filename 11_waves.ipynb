```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
def process(func, steps):
    # tabulate values
    data = [func(t / steps) for t in range(int(2 * steps * math.pi) + 1)]

    # plot the original function
    plt.plot(data)
    plt.show()

    # calculate spectrum
    amp = np.fft.fft(data)

    # real part
    plt.bar(list(range(len(amp))), list(map(lambda x: x.real, amp)))
    plt.show()

    # imaginary part
    plt.bar(list(range(len(amp))), list(map(lambda x: x.imag, amp)))
    plt.show()

    # absolute values
    plt.bar(list(range(len(amp))), list(map(lambda x: abs(x), amp)))
    plt.show()

    # inverse transform
    data = np.fft.ifft(amp)

    # plot recovered data
    plt.plot(list(map(lambda x: x.real, data)))
    plt.show()
    #plt.plot(list(map(lambda x: x.imag, data)))
    #plt.show()
```

```python
# e^i(x) = cos(x) + i*sin(x)

process(lambda x: math.sin(x), 20)
```

```python
process(lambda x: math.sin(3 * x) + math.sin(7 * x), 10)
```

```python
process(lambda x: math.sin(3 * x) - 2 * math.sin(7 * x), 10)
```

```python
process(lambda x: math.sin(3 * x) - math.cos(7 * x), 10)
```

```python
process(lambda x: math.sin(x) + 1, 10)
```

```python
# sin(a+b) = sin(a)cos(b) + cos(a)*sin(b)

process(lambda x: math.sin(x + math.pi / 4), 10)
```
