```python
import numpy as np

def one_position(position, d, i):
    # sin(pos / 10000 ^ (i / d))
    # cos(pos / 10000 ^ (i / d))
    # d - dimension of the positional encoding
    # i - ith component of the result
    # pos - position of the word in the sequence
    if i % 2 == 0:
        return math.sin(position / (10000 ** (i / d)))
    else:
        return math.cos(position / (10000 ** ((i - 1) / d)))

def positional(position, d):
    return np.array([one_position(position, d, i) for i in range(0, d, 1)])

plt.plot(positional(5, 100))
plt.show()

plt.plot(positional(80, 100))
plt.show()

plt.plot(positional(170, 100))
plt.show()
```

```python
cax = plt.matshow([positional(i, 100) for i in range(150)])
plt.gcf().colorbar(cax)
```

```python
for i in range(32):
    print("{:>3} ".format(i) + " ".join(map(lambda x: "{:>8.5f}".format(x), positional(i, 10))))
```

```python
for k in range(10):
    print(list(filter(lambda x: positional(x, 10)[k] > 0.9, range(32))))
```

```python
for k in range(10):
    print(list(filter(lambda x: positional(x, 10)[k] < -0.9, range(32))))
```
